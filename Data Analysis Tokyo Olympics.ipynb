```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
athletes_df = pd.read_csv("C:\\Computer Science\\Data Analytics\\Tokyo Olympics\\Athletes.csv")
gender_df = pd.read_csv("C:\\Computer Science\\Data Analytics\\Tokyo Olympics\\EntriesGender.csv")
medal_df = pd.read_csv("C:\\Computer Science\\Data Analytics\\Tokyo Olympics\\Medals.csv")
coaches_df = pd.read_csv("C:\\Computer Science\\Data Analytics\\Tokyo Olympics\\Coaches.csv")
team_df = pd.read_csv("C:\\Computer Science\\Data Analytics\\Tokyo Olympics\\Teams.csv")
```

```python
gender_df.head()
```

```python
gender_df.isnull().sum()
# NO NULL Value is present in this dataframe
```

```python
male_participant = gender_df['Male'].sum()
female_participant = gender_df['Female'].sum()
```

```python
data = np.array([male_participant,female_participant])
plt.pie(data,labels = [f"Male Participant are {male_participant}",f"Female Participants {female_participant}"],autopct = "%1.1f%%");
plt.title("Male And Female Participant")
plt.rcParams['figure.figsize'] = (16,8)
plt.rcParams['figure.facecolor'] = '#FFE5B4'
```

```python
# Analysing from which countries most of the athletes has come from
athletes_df
```

```python
countries = athletes_df['NOC'].value_counts().index[0:20]
athletes_count = athletes_df['NOC'].value_counts().values[0:20]
plt.rcParams['figure.figsize'] = (30,10)
plt.title("Countries with their Total number of athletes")
plt.xlabel("Countries")
plt.ylabel("No. of Athletes")
plt.xticks(rotation = 30,ha = "right")
plt.bar(countries,athletes_count);
```

```python
# Now we will see which discipline is most popular and which country has the highest participant in it
popu_discipline = athletes_df['Discipline'].value_counts().index[0:10]
popu_discipline_value = athletes_df['Discipline'].value_counts().values[0:10]
```

```python
print(f"Most Athletes come from {countries[0]} which is {athletes_count[0]}")
```

```python
team_df.isnull().sum()
```

```python
a = team_df[team_df['Discipline']=='Athletics']['Name'].value_counts().index[0]
b = team_df[team_df['Discipline']=='Athletics']['Name'].value_counts().values[0]
print(f"Most popular Discipline is {popu_discipline[0]} and most athletes in this discipline comes from {a} which is {b}")
```

```python
plt.rcParams['figure.figsize'] = (20,8)
plt.title("Most popular Discipline")
plt.xlabel("Discipline")
plt.ylabel("No. Of Athethes")
plt.xticks(rotation = 30,ha = "right")
plt.bar(popu_discipline,popu_discipline_value);
```

```python
coaches_df.isnull().sum()
```

```python
# Now will see which countries produces the most coaches
countris_coaches = coaches_df['NOC'].value_counts().index[0:5]
countris_coaches_count = coaches_df['NOC'].value_counts().values[0:5]
data = np.array(countris_coaches_count)
plt.pie(data,labels=np.array(countris_coaches),autopct = "%1.1f%%");
```

```python
# Countries which receive most gold deals
sorted_g = medal_df.sort_values("Gold",ascending = False)
plt.rcParams['figure.figsize'] = (30,10)
plt.xlabel("Countries")
plt.ylabel("No. of Gold medals")
plt.title("Countries with most number of Gold medals")
plt.xticks(rotation = 90,ha = "right")
plt.bar(sorted_g['Team/NOC'][0:20],sorted_g['Gold'][0:20]);
```

```python
# countries with most number of Silver Medals
sorted_s = medal_df.sort_values("Silver",ascending = False)
plt.rcParams['figure.figsize'] = (30,10)
plt.xlabel("Countries")
plt.ylabel("No. of Silver medals")
plt.title("Countries with most number of silver medals")
plt.xticks(rotation = 90,ha = "right")
plt.bar(sorted_s['Team/NOC'][0:20],sorted_s['Silver'][0:20]);
```

```python
# Countries with most number of bronze medals
sorted_b = medal_df.sort_values("Bronze",ascending=False)
plt.rcParams['figure.figsize'] = (30,10)
plt.xlabel("Countries")
plt.ylabel("No. of Bronze medal")
plt.title("Countries with most number of Bronze medals")
plt.xticks(rotation = 90,ha = "right")
plt.bar(sorted_b['Team/NOC'][0:20],sorted_b['Bronze'][0:20]);
```

```python
# Analysing One particular Country
g_i = sorted_g[sorted_g['Team/NOC'] =='India'].iloc[0]['Gold']
s_i = sorted_g[sorted_g['Team/NOC'] =='India'].iloc[0]['Silver']
b_i = sorted_g[sorted_g['Team/NOC'] =='India'].iloc[0]['Bronze']
data = np.array(["Gold","silver","Bronze"])
medal_c_i = np.array([g_i,s_i,b_i])
plt.title("India with Number of Medals it Won")
plt.xlabel("Medals")
plt.ylabel("Medal Counts")
plt.rcParams['figure.figsize'] = (12,8)
plt.bar(data,medal_c_i);
```
